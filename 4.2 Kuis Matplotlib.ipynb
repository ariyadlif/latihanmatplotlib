```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Soal 1
Dari data x dan y buatlah sebuah figure dengan ketentuan :
1. Judul figure adalah Data Penjualan PT. Sejahtera
2. X label adalah Pembelian
3. Y label adalah penjualan

```python
x = [0,20,30,45,60,100]
y = [0,40,60,90,120,200]

plt.plot(x,y)
plt.xlabel('Pembelian')
plt.ylabel('Penjualan')
print("Text(0.5, 1.0, 'Data Penjualan PT.Sejahtera')")
plt.title('Data Penjualan PT.Sejahtera')
plt.show()
```

```python
x = np.linspace(0,2*np.pi, 100)
y = x ** 2

plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.plot(x,y, 'b*-')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0,2*np.pi, 100)

r = np.sqrt(20.0)
r2 = np.sqrt(15.0)
r3 = np.sqrt(10.0)
r4 = np.sqrt(5.0)

x1 = r*np.cos(theta)
x2 = r*np.sin(theta)

x3 = r2*np.cos(theta)
x4 = r2*np.sin(theta)

x5 = r3*np.cos(theta)
x6 = r3*np.sin(theta)

x7 = r4*np.cos(theta)
x8 = r4*np.sin(theta)

fig, ax = plt.subplots(1)

ax.plot(x1, x2, x3, x4,'--r', x5, x6,'--*g', x7, x8,'y')
ax.set_aspect(1)


plt.xlim(-5, 5)
plt.ylim(-5, 5)


plt.show()
```
